```python
def matrixFromFile(filename):
    file = open(filename,"r")
    text = file.read()
    rows = re.split("NEW_ROW,", text)
    matrix = []
    for row in rows[:-1]:
        matrix.append(re.split(",", row)[:-1])
    matrix = np.array(matrix)
    matrix = matrix.astype(np.float)
    return matrix
```

```python
def getPublishedMatrices(filename):
    bag = rosbag.Bag(filename)
    matrices = []
    for topic, msg, t in bag.read_messages(topics=["/roboy_dep/depLoadMatrix"]):
        matrix = []
        m = msg.depMatrix
        for i in range(len(m)):
            matrix.append(m[i].cArray)
        matrices.append(np.array(matrix))
    bag.close()
    return matrices
```
